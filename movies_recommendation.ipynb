# Task - 2 NLP Text preprocessing.

```python
# importing libraries.
import nltk
from nltk.corpus import stopwords
from nltk.stem import WordNetLemmatizer
import re
```

```python
# both are help in tokenization.
# now donwload the 'stopword & wordnet'

nltk.download('stopwords')
nltk.download('wordnet')
```

```python
# remove all the stopwords
stop_words = set(stopwords.words('english'))
lemmatizer = WordNetLemmatizer()
```

```python
def preprocess_text(text):
  #lowercase
  text = str(text).lower()
  #removed the punctuations
  text = re.sub(r'[^a-zA-Z\s]', '', text)

  words = text.split()
  #remove stopwords
  words = [word for word in words if word not in stop_words]

  #lemmatize
  words = [lemmatizer.lemmatize(word) for word in words]

  return " ".join(words)
```

```python
# apply this function on 'tags' column.
# this function help as remove all the stopword 
# from these columns (overview, genres, tagline)
# this is the reason i'm create a new feature called "tags"

df['tags'] = df['tags'].apply(preprocess_text)
```

```python
# now i show the result of this.
# all words in lower case & stopwords are remove.

df['tags'][0]
```

```python
# reset all the data.
df = df.reset_index(drop = True)
```

```python
# now all movies have a 'indices'
indices = pd.Series(df.index,index = df['title']).drop_duplicates()
indices
```

# Task 3 - Text Vectorization.

```python
# importing 'TF-IDF'

from sklearn.feature_extraction.text import TfidfVectorizer
```

```python
tfidf = TfidfVectorizer(max_features=50000,ngram_range=(1,2),stop_words='english')
```

```python
tfidf_matrix = tfidf.fit_transform(df['tags'])
```

```python
# see the matrix.
tfidf_matrix
```

### Cosine Similarity.
#### Recommendation Systems: It helps identify items or users with similar preferences by comparing their feature vectors.

```python
# Cosine Similarity
from sklearn.metrics.pairwise import cosine_similarity
```

```python
"""
Create a function to 
find the most similar movies based 
on text features (like plot descriptions, genres, 
or other metadata encoded in tfidf_matrix).
"""

def recommend(title, n = 10):
  if title not in indices:
    return ['Movie not found']

  idx = indices[title]
  sim_score = cosine_similarity(tfidf_matrix[idx], tfidf_matrix).flatten()
  similar_idx = sim_score.argsort()[::-1][1:n+1]
  return df['title'].iloc[similar_idx]
```

```python
# lets see the recommendtions.

recommend('Toy Story')
```

```python
"""
importing pickle for 
building a movie recommendation (UI)
"""

import pickle

pickle.dump(tfidf_matrix,open('tfidf_matrix.pkl','wb'))

pickle.dump(indices,open('indices.pkl','wb'))

df.to_pickle('df.pkl')

pickle.dump(tfidf,open('tfidf.pkl','wb'))
```
